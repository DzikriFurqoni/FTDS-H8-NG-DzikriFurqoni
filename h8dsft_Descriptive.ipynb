```python
obj_cols= list(df.select_dtypes('object'))
num_cols= list(df.select_dtypes('number'))
print(obj_cols)
print(num_cols)
```

```python
df[num_cols].describe()
```

```python
df[obj_cols].describe()
```

```python
# Mencari mean median and mode

data_mean = df[num_cols].mean().rename('mean')
data_median = df[num_cols].median().rename('median')
data_mode = df[num_cols + obj_cols].mode().iloc[0].rename('mode')

central_tendency = pd.concat([data_mean, data_median, data_mode], axis=1)
central_tendency
```

```python

data_var = df[num_cols].var().rename('variance')
data_std = df[num_cols].std().rename('standart_deviaton')

var_std = pd.concat([data_var, data_std], axis=1)
var_std
```
